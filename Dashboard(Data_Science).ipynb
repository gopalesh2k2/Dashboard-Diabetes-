```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
#reading our dataset and storing it as a pabndas data frame
df = pd.read_csv('diabetes-1.csv')
cols = df.columns
df.head()
```

```python
# it gives us all the info about what is there in our data set all the columns and their data data types and number of non null values
print(df.info())
```

```python
#this will give us the null values in each of our column
print(df.isnull().sum())
#since their are no null values we dont need to apply any data filling procedures
```

```python
# to understand how our data population is divided
sns.countplot(x ='Outcome', data = df,palette='Blues').set(title='Count of Non-Diabetic and Diabetic Patients ')
plt.show()
# this shows most people are non diabetic
```

#Relationship


Scatter Plot

```python
# since we have 9 effective features drawing a scatter plot for all is not a viable option
# so we wanted to give our user the option to be able to choose the features between which they want to make analysis
# a scatter plot would allow the user to be able to understand the value distribution and infer regions of interest
print("Select any of the following mentioned features to draw a SCATTER PLOT:")
for i in range(len(cols)):
  print("{}. {}".format(i,cols[i]))
print()
x = int(input('Select X axis feature(index no):'))
y = int(input('select Y axis feature(index no):'))
sns.scatterplot(x = cols[x], y = cols[y], data = df, hue='Outcome').set(title=cols[y]+" v/s "+cols[x])
plt.show()
```

Correlation Plot

```python
# whenever we want to train a ML model or if we want to analyse the data 
# the analysis is easier if we are able to cut down to interest points the correlation plot 
# tells about how features are interrelated and we can select the features that suits us most
plt.figure(figsize=(10,7))
sns.heatmap(df.corr(), annot = True).set(title = 'Correlation Plot')
plt.show()
```

#Distribution

Box Plot and Histogram

```python
# the histograms are used here to understand how a particular feature is distributed
# tells us about the count of each feature group value

# box plot is used to visualise one or many distributions in a dataset
# box plot divides tye data into quartiles
# it gives info about skewness of data 
# it marks otlier values
plt.figure(figsize=(25,45))
pos = 1
for i in range(0,8):
  plt.subplot(8,2,pos)
  sns.boxplot(x = 'Outcome', y = cols[i], data = df).set(title = cols[i]+' Analysis')
  pos = pos+1
  plt.subplot(8,2,pos)
  sns.histplot(data = df, x = cols[i],hue= 'Outcome').set(title = cols[i] + ' Distribution')
  pos = pos+1
plt.show()
```
